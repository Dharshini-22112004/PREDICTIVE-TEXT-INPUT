```python
from transformers import pipeline

def predict_next_word(input_text):
    # Load the text generation pipeline
    generator = pipeline('text-generation', model='gpt2')  # You can choose other models as well

    # Generate predictions based on the input text
    predictions = generator(input_text, max_length=len(input_text.split()) + 5, num_return_sequences=3)

    # Extract the predicted text
    for prediction in predictions:
        print(prediction['generated_text'])

# Test the function
input_text = "The weather today is"
print(f"Input: {input_text}")
print("Predictions:")
predict_next_word(input_text)
```

```python
# A loop to allow for multiple inputs
inputs = ["The weather today is", "I am planning to go", "Artificial Intelligence is"]

for input_text in inputs:
    print(f"Input: {input_text}")
    print("Predictions:")
    predict_next_word(input_text)
```
